##### __Definindo Localiza√ß√£o__

```python
import folium
```

```python
#Latitude e longitude - Jo√£o Pessoa

#latitude = -7.5555
latitude = -7.11532
longitude = -34.861

#Raio em kil√¥metros
raio = 18
```

```python
Paraiba = 'PB.json'

mapa = folium.Map(
    location = [latitude, longitude],
    zoom_start=11
)

folium.GeoJson(
    Paraiba,
    name='geojson'
).add_to(mapa)

folium.Circle(
    radius=raio*1000,
    location=[latitude, longitude],
    popup='Centro',
    color='crimson',
    fill=False,
).add_to(mapa)
mapa
```

```python
#string de localiza√ß√£o
geocodeStr = '{},{},{}km'.format(latitude, longitude, raio)
geocodeStr
```

```python
locais = api.reverse_geocode(lat = latitude,long = longitude, accuracy='18000m')
[print(local.full_name) for local in locais ]
```

```python
chuvas = api.search(q = 'chuva OR enchente OR chuvendo OR chove',
                    geocode = geocodeStr, 
                    count = 101,
                    tweet_mode = 'extended')
```

```python
len(chuvas)
```

```python
[print(tweet.full_text) for tweet in chuvas]
```

```python
[print(tweet.user.location) for tweet in chuvas]
```

```python
[print(tweet.created_at) for tweet in chuvas]
```

```python
palavras = {'chuva':0, 'enchente':0, 'outro':0}

for tweet in chuvas:
    if 'chuva' in tweet.text.lower():
        palavras['chuva'] = palavras['chuva'] +1
    if 'enchente' in tweet.text.lower(): 
        palavras['enchente'] = palavras['enchente'] +1
        print(tweet.text)
    if 'chuva' not in tweet.text.lower() and 'enchente' not in tweet.text.lower():
        palavras['outro'] = palavras['outro'] + 1
        print(tweet.text)
        
        
```
